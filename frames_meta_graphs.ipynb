```python
import json
import pandas as pd

import matplotlib.pyplot as plt
```

```python
with open('/Users/elhidze/Gagarin/metadata/train/3.json', 'r') as file:
    frames = json.load(file)

frames_json = json.dumps(frames)

frames = json.loads(frames_json)

pkt_pos = [int(frame['pkt_pos']) for frame in frames]
pkt_pos = [i / 25 / 60 for i in range(len(pkt_pos))]
pkt_sizes = [int(frame['pkt_size']) / frame['height'] / frame['width'] for frame in frames]
pic_types = [frame['pict_type'] for frame in frames]

I_frames_pos = [pos for pos, pic_type in zip(pkt_pos, pic_types) if pic_type == 'I']
I_frames_sizes = [size for size, pic_type in zip(pkt_sizes, pic_types) if pic_type == 'I']
P_frames_pos = [pos for pos, pic_type in zip(pkt_pos, pic_types) if pic_type == 'P']
P_frames_sizes = [size for size, pic_type in zip(pkt_sizes, pic_types) if pic_type == 'P']
B_frames_pos = [pos for pos, pic_type in zip(pkt_pos, pic_types) if pic_type == 'B']
B_frames_sizes = [size for size, pic_type in zip(pkt_sizes, pic_types) if pic_type == 'B']

# Create timeline plot with different colors for each frame type
plt.scatter(I_frames_pos, I_frames_sizes, color='red', marker='o', label='I Frames')
plt.scatter(P_frames_pos, P_frames_sizes, color='green', marker='o', label='P Frames')
plt.scatter(B_frames_pos, B_frames_sizes, color='blue', marker='o', label='B Frames')

plt.title('Packet Size Distribution Over Time')
plt.xlabel('Packet Position')
plt.ylabel('Packet Size')
plt.legend()
plt.grid(True)
plt.show()
```
